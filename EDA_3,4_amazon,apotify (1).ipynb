```python
#Que 1 What is the average rating for each product category?


import pandas as pd

# Step 1: Load the Dataset
df = pd.read_csv("amazon.csv")

# Step 2: Extract Main Category
df['main_category'] = df['category'].astype(str).str.split('|').str[0]

# Step 3: Convert Rating to Numeric
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Step 4: Calculate Average Rating per Category
avg_rating = df.groupby('main_category')['rating'].mean().reset_index()

# Step 5: Sort by Rating (Descending)
avg_rating = avg_rating.sort_values(by='rating', ascending=False)

# Step 6: Show Result
print(" Average Rating by Product Category:\n")
print(avg_rating)
```

```python
#Que 2 What are the top rating_count products by category?

import pandas as pd

# Step 1: Load the dataset
df = pd.read_csv("amazon.csv")

# Step 2: Clean and convert 'rating_count'
df['rating_count'] = df['rating_count'].astype(str).str.replace(',', '')
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')

# Step 3: Extract main category
df['main_category'] = df['category'].astype(str).str.split('|').str[0]

# Step 4: Find top rating_count product in each category
top_products = df.loc[df.groupby('main_category')['rating_count'].idxmax()][
    ['main_category', 'product_name', 'rating_count']
].sort_values(by='rating_count', ascending=False)

# Step 5: Show Result
print("📊 Top Rating Count Products by Category:\n")
print(top_products)
```

```python
#Que 3  What is the distribution of discounted prices vs. actual prices?

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load the dataset
df = pd.read_csv("amazon.csv")

# Step 2: Clean price columns
df['discounted_price'] = df['discounted_price'].astype(str).str.replace('₹', '').str.replace(',', '')
df['actual_price'] = df['actual_price'].astype(str).str.replace('₹', '').str.replace(',', '')

# Step 3: Convert to numeric
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')
df['actual_price'] = pd.to_numeric(df['actual_price'], errors='coerce')

# Step 4: Plot distribution
plt.figure(figsize=(10, 6))
sns.kdeplot(df['actual_price'].dropna(), label='Actual Price', fill=True)
sns.kdeplot(df['discounted_price'].dropna(), label='Discounted Price', fill=True)
plt.title("Distribution: Discounted vs Actual Prices")
plt.xlabel("Price (₹ INR)")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Que 4 How does the average discount percentage vary across categories?

import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the dataset
df = pd.read_csv("amazon.csv")

# Step 2: Extract main category
df['main_category'] = df['category'].astype(str).str.split('|').str[0]

# Step 3: Clean and convert discount_percentage
df['discount_percentage'] = df['discount_percentage'].astype(str).str.replace('%', '')
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'], errors='coerce')

# Step 4: Calculate average discount by category
avg_discount = df.groupby('main_category')['discount_percentage'].mean().sort_values(ascending=False)

# Step 5: Plot the result
avg_discount.plot(kind='bar', figsize=(10, 5), color='orange')
plt.title("Average Discount Percentage by Category")
plt.ylabel("Average Discount (%)")
plt.xlabel("Product Category")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Que 5 What are the most popular product names?

import pandas as pd

# Step 1: Load the dataset
df = pd.read_csv("amazon.csv")

# Step 2: Clean and convert rating_count
df['rating_count'] = df['rating_count'].astype(str).str.replace(',', '')
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')

# Step 3: Sort by rating_count descending
popular_products = df[['product_name', 'rating_count']].sort_values(by='rating_count', ascending=False)

# Step 4: Show top 10
print("📦 Top 10 Most Popular Products (by Rating Count):\n")
print(popular_products.head(10))
```

```python
#Que 6 What are the most popular product keywords?


import pandas as pd
from collections import Counter
import re

# Step 1: Load the dataset
df = pd.read_csv("amazon.csv")

# Step 2: Combine all product names
all_names = ' '.join(df['product_name'].dropna().astype(str))

# Step 3: Tokenize and clean words
words = re.findall(r'\b\w+\b', all_names.lower())  # Convert to lowercase and remove symbols

# Step 4: Remove common stopwords
stopwords = set(['and', 'for', 'with', 'the', 'of', 'in', 'on', 'by', 'to', 'at', 'from', 'a', 'an'])
filtered_words = [word for word in words if word not in stopwords and len(word) > 2]

# Step 5: Count most common words
keyword_counts = Counter(filtered_words).most_common(20)

# Step 6: Display top keywords
print("🔍 Top 20 Product Keywords:\n")
for keyword, count in keyword_counts:
    print(f"{keyword}: {count}")
```

```python
#Que 7 What are the most popular product reviews?


import pandas as pd
from collections import Counter

# Step 1: Load the dataset
df = pd.read_csv("amazon.csv")

# Step 2: Drop missing reviews
df_reviews = df['review_content'].dropna()

# Step 3: Count most common reviews
review_counts = Counter(df_reviews).most_common(10)

# Step 4: Display top reviews
print("📝 Top 10 Most Frequent Reviews:\n")
for i, (review, count) in enumerate(review_counts, start=1):
    print(f"{i}. ({count} times)\n{review}\n")
```

```python
#Que 8 What is the correlation between discounted_price and rating?

import pandas as pd

# Load the dataset
df = pd.read_csv("amazon.csv")

# Clean discounted_price and rating columns
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Drop rows where either value is NaN
df_corr = df[['discounted_price', 'rating']].dropna()

# Correlation calculation
correlation = df_corr['discounted_price'].corr(df_corr['rating'])

# Print result
print("📊 Correlation between Discounted Price and Rating:", round(correlation, 3))

```

```python


import pandas as pd
# Step 1: Read the CSV file
df = pd.read_csv ("spotify.csv")  # Replace with your actual file name

# Step 2: Check for null values
print("Null values in each column:\n", df.isnull().sum())

# Step 3: Drop rows with null values (if any)
df = df.dropna()

# Step 4: Check for duplicate rows
print("\nNumber of duplicate rows:", df.duplicated().sum())

# Step 5: Drop duplicate rows (if any)
df = df.drop_duplicates()

# Step 6: Display cleaned dataset
df.head()

```

```python
#Que 2 What is the distribution of popularity among the tracks in the dataset? Visualize it using a histogram?
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for the plot
sns.set(style="whitegrid")

# Create a histogram of the 'Popularity' column
plt.figure(figsize=(10, 6))
sns.histplot(df['Popularity'], bins=20, kde=True, color='purple')

# Add labels and title
plt.title('Distribution of Track Popularity', fontsize=16)
plt.xlabel('Popularity Score')
plt.ylabel('Number of Tracks')

# Show the plot
plt.show()
```

```python
#Que 3 Is there any relationship between the popularity and the duration of tracks? Explore this using a scatter plot?

import matplotlib.pyplot as plt
import seaborn as sns

# Set the style
sns.set(style="darkgrid")

# Create the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Duration (ms)', y='Popularity', alpha=0.6, color='teal')

# Add labels and title
plt.title('Popularity vs Duration of Tracks', fontsize=16)
plt.xlabel('Duration (ms)')
plt.ylabel('Popularity Score')

# Show the plot
plt.show()
```

```python
#Que 4 Which artist has the highest number of tracks in the dataset? Display the count of tracks for each artist using
# a countplot?

import matplotlib.pyplot as plt
import seaborn as sns

# Set the plot style
sns.set(style="whitegrid")

# Count the number of tracks per artist
plt.figure(figsize=(14, 6))
top_artists = df['Artist'].value_counts().head(10)  # Top 10 artists for better readability

# Plot using Seaborn countplot
sns.countplot(data=df[df['Artist'].isin(top_artists.index)], y='Artist', order=top_artists.index, palette='viridis')

# Add labels and title
plt.title('Top 10 Artists with the Most Tracks', fontsize=16)
plt.xlabel('Number of Tracks')
plt.ylabel('Artist')

# Show plot
plt.tight_layout()
plt.show()
```

```python
#Que 5 What are the top 5 least popular tracks in the dataset? Provide the artist name and track name for each?

# Sort the dataframe by popularity (ascending)
least_popular_tracks = df.sort_values(by='Popularity', ascending=True).head(5)

# Select relevant columns: Artist and Track Name
least_popular_tracks_info = least_popular_tracks[['Artist', 'Track Name', 'Popularity']]

# Display the result
print("Top 5 Least Popular Tracks:")
print(least_popular_tracks_info)
```

```python
#Que 6  Among the top 5 most popular artists, which artist has the highest popularity on average? Calculate and
#. display the average popularity for each artist?

# Step 1: Group by artist and calculate total popularity
artist_total_popularity = df.groupby('Artist')['Popularity'].sum().sort_values(ascending=False)

# Step 2: Select top 5 artists based on total popularity
top_5_artists = artist_total_popularity.head(5).index

# Step 3: Filter original dataframe for only top 5 artists
top_5_df = df[df['Artist'].isin(top_5_artists)]

# Step 4: Calculate average popularity per artist
artist_avg_popularity = top_5_df.groupby('Artist')['Popularity'].mean().sort_values(ascending=False)

# Step 5: Display results
print("Average Popularity of Top 5 Most Popular Artists:")
print(artist_avg_popularity)
```

```python
# Que 7  For the top 5 most popular artists, what are their most popular tracks? List the track name for each artist?

import pandas as pd

# Step 1: Load the CSV (if not already done)
df = pd.read_csv('spotify.csv')

# Step 2: Get top 5 most popular artists based on average popularity
top_5_artists = df.groupby('Artist')['Popularity'].mean().sort_values(ascending=False).head(5).index

# Step 3: Filter only those top 5 artists
top_5_df = df[df['Artist'].isin(top_5_artists)]

# Step 4: Find the most popular track for each of the top 5 artists
top_tracks = top_5_df.loc[
    top_5_df.groupby('Artist')['Popularity'].idxmax(),
    ['Artist', 'Track Name', 'Popularity']
]

# Step 5: Display result
print("🎧 Most popular track for each of the top 5 most popular artists:")
print(top_tracks)



```

```python
#Que 9 Does the duration of tracks vary significantly across different artists? Explore this visually using a box plot or
#   violin plot?

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load data
df = pd.read_csv('spotify.csv')

# Step 2: Select top 5 most popular artists for cleaner visualization
top_5_artists = df.groupby('Artist')['Popularity'].mean().sort_values(ascending=False).head(5).index
top_df = df[df['Artist'].isin(top_5_artists)]

# Step 3: Convert duration from milliseconds to minutes (optional)
top_df['Duration (min)'] = top_df['Duration (ms)'] / 60000

# Step 4: Create box plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=top_df, x='Artist', y='Duration (min)', palette='pastel')
plt.title('Track Duration Distribution for Top 5 Artists')
plt.ylabel('Duration (minutes)')
plt.xlabel('Artist')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(data=top_df, x='Artist', y='Duration (min)', palette='muted')
plt.title('Track Duration Spread per Artist (Top 5)')
plt.ylabel('Duration (minutes)')
plt.xlabel('Artist')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
